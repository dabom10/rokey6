```python
arr = [3, 1, 2]

arr.sort()         # ✅ 원본이 정렬됨
print(arr)         # [1, 2, 3]

arr = [3, 1, 2]
new_arr = sorted(arr)  # ✅ 새로운 리스트를 반환
print(arr)             # [3, 1, 2] (원본 유지)
print(new_arr)         # [1, 2, 3]
```

```python
# np.sort(), arr.sort() 차이 : 원본 수정 여부
import numpy as np
arr = np.array([[3, 1, 2],
                [9, 8, 7]])

print(np.sort(arr, axis=1))
# 각 행(가로) 기준 정렬
# [[1, 2, 3],
#  [7, 8, 9]]

arr.sort(axis=0)  # 원본 배열이 열 단위로 정렬됨
print(arr)
# [[3, 1, 2],
#  [9, 8, 7]] → (axis=0이면 세로 기준 정렬)
```

```python
# 텐서는 항상 새로운 배열 반환
import torch

t = torch.tensor([[3, 1, 2],
                  [9, 8, 7]])

values, indices = t.sort(dim=1)
print(values)
# tensor([[1, 2, 3],
#         [7, 8, 9]])

print(indices)
# tensor([[1, 2, 0],
#         [2, 1, 0]])  ← 원래 위치
```
