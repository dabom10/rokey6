ì‘ìš© ì˜ˆì‹œ

```python
# ë‘ ë³€ìˆ˜ë¥¼ ë¹„êµí•´ì„œ ì¼ì¹˜í•˜ëŠ” ê°œìˆ˜ë¡œ ì •í™•ë„ ê³„ì‚°
# 2ê°œì˜ ë„˜íŒŒì´ ë°°ì—´ ì¤€ë¹„

# ì •ë‹µ
yt = np.array([1,1,0,1,1,0,1,1,0,0])

# ì˜ˆì¸¡
yp = np.array([1,1,0,1,1,1,0,0,1,1])
```

```python
# ì •ë‹µ(y)ê³¼ ì˜ˆì¸¡(y_pred) ì˜ ë§ë‚˜? (ë™ì‹œë¹„êµ)

matched = (yt == yp)
print(matched)
```

```python
matched.sum() # True = 1ì´ë‹ˆê¹Œ
```

```python
# ë§ì€ ê°œìˆ˜
correct = matched.sum()
# ì „ì²´ ê°œìˆ˜
total = len(matched)
```

```python
matched.mean()
```

```python
# ì •í™•ë„ ê³„ì‚°
accuarcy = correct / total
print(accuarcy)
print(f'''
      ì •ë‹µ ìˆ˜: {correct},
      ì „ì²´ ìˆ˜: {total},
      ì •í™•ë„ : {accuarcy: .3f} # í¬ë§· í•¨ìˆ˜ ì ìš©
      ''')
```
