```python
import torch

random_tensor = torch.randn((2,3))

print(random_tensor)

tensor_like = torch.randn_like(random_tensor)

print(tensor_like) #동일크기 동일데이터타입 -1~+1 랜덤
```

```python
import torch

exam = torch.rand(3,1,4,1)
print(exam)
exam = exam.squeeze(dim=1)
print(exam)
exam = exam.unsqueeze(dim=0)
print(exam)
```

```python
import torch

x = torch.tensor([[[1], [2], [3]]])
print(x)
print(x.shape) # torch.Size([1, 3, 1])
print(x.squeeze())
print(x.squeeze().shape) # torch.Size([3])
```

```python
import torch

x = torch.tensor([[[1], [2], [3]]])
print(x)
print(x.shape) # torch.Size([1, 3, 1])
print(x.squeeze(dim=0))
print(x.squeeze(dim=0).shape) # torch.Size([3, 1])
```
